```python
X = data.iloc[:,:-1] # independent variables
y = data["1"] # target column or dependent variable
```

#### Transforming values of X using standard scaler

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X)
X=sc.transform(X)
```

## Perceptron Algorithm

```python
# Implementing Perceptron Algorithm
from sklearn.linear_model import Perceptron
model = Perceptron()
model.fit(X, y)
```

```python
#importing libraries for evaluating the perfomance

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from numpy import mean
from numpy import std
```

```python
#crosss validation
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)

#model evaluation
scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
print('Mean Accuracy: ',(mean(scores)))
```

```python
# Splitting the data set into two , train and test with a ratio (0.8,0.2)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=1)

#fitting model with train data
model.fit(X_train,y_train)
```

```python
#predcting y values with x_test

y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score #calculating the accuracy 
print('Accuracy: ',accuracy_score(y_test, y_pred))
```

```python
#Tune Perceptron Hyperparameters-Learning Rate

from sklearn.model_selection import GridSearchCV

grid = dict()
grid['eta0'] = [0.0001, 0.001, 0.01, 0.1, 1.0]
search = GridSearchCV(model, grid, scoring='accuracy', cv=cv, n_jobs=-1)
results = search.fit(X,y)

results
```

```python
# evaluation
print(results.best_score_)
print(results.best_params_)
```

Model performed well with smaller learning rates .0001 , model achieved best accuracy

```python
# Epoch Rate

grid = dict()
grid['max_iter'] = [1, 10, 100, 1000, 10000]
re_model = Perceptron(eta0=0.0001)
search = GridSearchCV(re_model, grid, scoring='accuracy', cv=cv, n_jobs=-1)
results = search.fit(X,y)
print(results.best_score_)
print( results.best_params_)
```

Epoch rate and best score is obtained

```python
#fitting the best model with tuned parameters

best_model=Perceptron(eta0=0.0001,max_iter=10)
best_model.fit(X_train,y_train)


y_pred = best_model.predict(X_test)
print('Accuracy:',accuracy_score(y_test, y_pred))
```

```python
#fitting best model

model_tuned = Perceptron(eta0=0.0001,max_iter=10)
model_tuned.fit(X,y)

#model evaluation
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
scores = cross_val_score(model_tuned, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
print(mean(scores))
```
